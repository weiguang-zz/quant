```python
#删选方法，K在20左右向上交叉D时，视为买进信号
# k是rsv（未成熟随机指标值）的指数移动平均线，d是k的指数移动平均线。rsv小于50时表示收盘价
# 更加接近于最低价，所以k在20左右向上交叉d表示此时价位较低（9日来看的话） 且具有增长的动量。
alllen = len(config.codes)
count = 0
choose=[]
for code in config.codes:
    code = code.replace('WIKI/','')
    firstc = code[0]
    sql = 'select * from prices_%s where code="%s"' % (firstc, code)
    daily_data=pd.read_sql(sql,db.session.connection())
    high = daily_data['High'].values
    low = daily_data['Low'].values
    close=daily_data['Close'].values
    k,d=talib.STOCH(high,low,close)
    df=pd.DataFrame(data={'k':k,'d':d},index=daily_data.Date)
    df['j']=3*df['k'] - 2*df['d']
    if df['k'][-1]<=25 and df['k'][-1] > df['d'][-1] \
         and df['k'][-2] < df['d'][-2]:
            choose.append(code)
            print '\n%s k线上穿d线' % code
    count+=1
    if count%10==0:
        sys.stdout.write("\r%.2f%%" % (float(count)/alllen*100))
        sys.stdout.flush()
```
